```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df =pd.read_csv('tested.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test =train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
x_train.head(2)
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
si_age=SimpleImputer()
si_fare =SimpleImputer(strategy='most_frequent')

x_train_age=si_age.fit_transform(x_train[['Age']])
x_train_fare=si_fare.fit_transform(x_train[['Fare']])

x_test_age=si_age.transform(x_test[['Age']])
x_test_fare=si_fare.transform(x_test[['Fare']])
```

```python
x_train_age
```

```python
ohe_sex = OneHotEncoder(sparse=False,handle_unknown ='ignore')
ohe_age = OneHotEncoder(sparse=False,handle_unknown ='ignore')

x_train_sex = ohe_sex.fit_transform(x_train[['Sex']])
x_train_age = ohe_age.fit_transform(x_train_age)

x_test_sex = ohe_sex.transform(x_test[['Sex']])
x_test_age = ohe_age.transform(x_test_age)
```

```python
x_train_age
```

```python
x_train.head(2)
```

```python
x_train_rem =x_train.drop(columns=['Sex','Age','Fare'])
```

```python
x_test_rem = x_test.drop(columns=['Sex','Age','Fare'])
```

```python
x_train_transformed =np.concatenate((x_train_rem,x_train_age,x_train_sex,x_train_fare),axis=1)
x_test_transformed =np.concatenate((x_test_rem,x_test_age,x_test_sex,x_test_fare),axis=1)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_transformed,y_train)
```
