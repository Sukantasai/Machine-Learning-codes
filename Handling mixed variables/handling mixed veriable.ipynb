```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with')
```

```python
#extract numerical part
df['number_numerical']=pd.to_numeric(df["number"],errors='coerce',downcast='integer')
```

```python
#extract categorical part
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['Cabin_num']=df['Cabin'].str.extract('(\d+)')#capture numerical part
df['Cabin_cat']=df['Cabin'].str[0]#capture the 1st leter
df.head()
```

```python
df['Cabin_cat'].value_counts().plot(kind='bar')
```

```python
#extract last bit of the ticket number
df['ticket_num']=df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num']=pd.to_numeric(df['ticket_num'],errors='coerce',
                              downcast='integer')
#extract the first part of the ticket category
df['ticket_cat']=df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat']=np.where(df['ticket_cat'].str.isdigit(),np.nan,
                         df['ticket_cat'])
df.head(20)
```
